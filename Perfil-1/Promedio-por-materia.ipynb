# Promedio de Cada Materia

<p>En esta sección se busca responder cúal es la materia de mayor y menor promedio a nivel nacional. Esto con el fin de reforzar la materia más difícil y a su vez, potenciar la materia donde es posible obtener puntaje perfecto.</p>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://www.datos.gov.co/resource/kgxf-xxbe.csv')
# df = pd.read_csv('/Users/Daniel\Desktop/programación/IntrodIngSistemasAnalíticaDeDatos/Resultados__nicos_Saber_11_20250316.csv')
```

## Promedio Matemáticas

## Promedio Matemáticas Base de Datos Completa

```python

promedios_por_periodo = df.groupby('periodo')['punt_matematicas'].mean()

tabla_promedios = df.groupby('periodo', as_index=False)['punt_matematicas'].mean()
tabla_promedios.rename(columns={'punt_matematicas': "promedio_matematicas"}, inplace=True)

plt.figure(figsize=(20, 16))
plt.bar(tabla_promedios['periodo'], tabla_promedios["promedio_matematicas"], color='skyblue')

plt.bar(tabla_promedios['periodo'], tabla_promedios["promedio_matematicas"], color=['orange', '#ffd000'])

plt.xlabel('periodo')  # Etiqueta para el eje X
plt.ylabel("Promedio Matemáticas")  # Etiqueta para el eje Y
plt.title("Promedio de Matemáticas por Periodo")  # Título de la gráfica

plt.xticks(tabla_promedios["periodo"])
promedios_por_periodo

plt.grid(axis="y", linestyle="--", alpha=0.7)
```

## Promedio Ciencias Naturales

## Promedio Ciencias Naturales Base de Datos Completa

```python
# Calcular promedio por periodo
promedios_por_periodo = df.groupby("periodo")["punt_c_naturales"].mean()

# Crear tabla de promedios por periodo
tabla_promedios = df.groupby("periodo", as_index=False)["punt_c_naturales"].mean()
tabla_promedios.rename(columns={"punt_c_naturales": "promedio_c_naturales"}, inplace=True)

# Graficar
plt.figure(figsize=(20, 16))
plt.bar(tabla_promedios["periodo"], tabla_promedios["promedio_c_naturales"], color=['#69c20f', '#95f733'])

plt.xlabel("Periodo")  # Etiqueta para el eje X
plt.ylabel("Promedio Ciencias Naturales")  # Etiqueta para el eje Y
plt.title("Promedio de Ciencias Naturales por Periodo")  # Título de la gráfica

plt.xticks(tabla_promedios["periodo"])
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

## Promedio Inglés

## Promedio Inglés Base de Datos Completa

```python
# Convertir la columna a numérica
df["punt_ingles"] = pd.to_numeric(df["punt_ingles"], errors="coerce")

# Calcular promedio por periodo
promedios_por_periodo = df.groupby("periodo")["punt_ingles"].mean()

# Crear tabla de promedios por periodo
tabla_promedios = df.groupby("periodo", as_index=False)["punt_ingles"].mean()
tabla_promedios.rename(columns={"punt_ingles": "promedio_ingles"}, inplace=True)

# Graficar
plt.figure(figsize=(20, 16))
plt.bar(tabla_promedios["periodo"], tabla_promedios["promedio_ingles"], color=['#15b293', '#10856e'])

plt.xlabel("Periodo")  # Etiqueta para el eje X
plt.ylabel("Promedio Inglés")  # Etiqueta para el eje Y
plt.title("Promedio de Inglés por Periodo")  # Título de la gráfica

plt.xticks(tabla_promedios["periodo"])
plt.grid(axis="y", linestyle="--", alpha=0.7)
(plt.xticks(rotation=45, ha="right"))

plt.show()
```

## Promedio Sociales y Ciudadanas

```python
df["punt_sociales_ciudadanas"] = pd.to_numeric(df["punt_sociales_ciudadanas"], errors="coerce")

# Calcular promedio por periodo
promedios_por_periodo = df.groupby("periodo")["punt_sociales_ciudadanas"].mean()

# Crear tabla de promedios por periodo
tabla_promedios = df.groupby("periodo", as_index=False)["punt_sociales_ciudadanas"].mean()
tabla_promedios.rename(columns={"punt_sociales_ciudadanas": "promedio_sociales_ciudadanas"}, inplace=True)

# Graficar
plt.figure(figsize=(20, 16))
plt.bar(tabla_promedios["periodo"], tabla_promedios["promedio_sociales_ciudadanas"], color=['#fff333', '#ffd000'])

plt.xlabel("Periodo")  # Etiqueta para el eje X
plt.ylabel("Promedio Sociales y Ciudadanas")  # Etiqueta para el eje Y
plt.title("Promedio de Sociales y Ciudadanas por Periodo")  # Título de la gráfica

plt.xticks(tabla_promedios["periodo"])
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

## Promedio Lectura Crítica

```python
df["punt_lectura_critica"] = pd.to_numeric(df["punt_lectura_critica"], errors="coerce")

# Calcular promedio por periodo
promedios_por_periodo = df.groupby("periodo")["punt_lectura_critica"].mean()

# Crear tabla de promedios por periodo
tabla_promedios = df.groupby("periodo", as_index=False)["punt_lectura_critica"].mean()
tabla_promedios.rename(columns={"punt_lectura_critica": "promedio_lectura_critica"}, inplace=True)

# Graficar
plt.figure(figsize=(20, 16))
plt.bar(tabla_promedios["periodo"], tabla_promedios["promedio_lectura_critica"], color=['#561dee', '#a081f1'])

plt.xlabel("Periodo")  # Etiqueta para el eje X
plt.ylabel("Promedio Lectura Crítica")  # Etiqueta para el eje Y
plt.title("Promedio de Lectura Crítica por Periodo")  # Título de la gráfica

plt.xticks(tabla_promedios["periodo"])
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

## Promedio de cada Materia de 2010 a 2022

```python
materias_a_promediar = ['punt_ingles', 'punt_matematicas', 'punt_sociales_ciudadanas', 'punt_c_naturales', 'punt_lectura_critica']

df[materias_a_promediar] = df[materias_a_promediar].replace(0, pd.NA)
promedio = df[materias_a_promediar].mean(skipna=True)
promedio.index = ["Inglés", "Matemáticas", "Ciencias Sociales", "Ciencias Naturales", "Lectura Crítica"]

plt.figure(figsize=(10, 6))
promedio.plot(kind='bar', color=['#4CAF50', '#2196F3', '#FF5722', '#9C27B0', '#FF9800'])

plt.title('Promedio por materia')
plt.ylabel('Promedio')
plt.xlabel('Materia')
plt.xticks(rotation=45, ha='right')
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

## Histograma de Cada Materia del 2010 a 2022

```python
text_dict = {'size': 16, 'color': '#581845'}

materias = ['punt_sociales_ciudadanas', 'punt_lectura_critica', 'punt_c_naturales', 'punt_matematicas', 'punt_ingles']
colores = ['#900C3F', '#FFC300', '#DAF7A6', '#FF5733', '#ff9999']
titulos = ['Puntaje Sociales y Ciudadanas', 'Puntaje Lectura Crítica', 
           'Puntaje Ciencias Naturales', 'Puntaje Matemáticas', 'Puntaje Inglés']

df_filtrado = df[materias].replace(0, pd.NA).dropna()

fig, axes = plt.subplots(1, 5, figsize=(16, 9))

for i, materia in enumerate(materias):
    axes[i].hist(df_filtrado[materia], color=colores[i])
    axes[i].set_xlabel(titulos[i], fontdict=text_dict)

plt.tight_layout()
plt.show()
```

## ¿Cuál es la materia con promedio más bajo y más alto?

<p>De acuerdo a las gráficas podemos notar que a lo largo del tiempo la materia con menor promedio es sociales y ciudadanas e inglés, sin embargo la estructura en el histograma de sociales es muy similar a la de lectura crítica y ciencias naturales por lo que vemos una tendencia a un promedio muy mantenido pero a pocos puntajes fuera de lo normal. Por otro lado, en Matemáticas y en Inglés pasa todo lo contrario, en el percentil se evidencia puntajes del máximo y una estructura muy similar.</p>

<p>Esto puede significar que, a pesar de la pequeña diferencia en promedio de inglés y sociales con las demás, realmente las materias más complicadas de obtener puntajes por encima de lo normal son sociales, español y naturales; a diferencia de inglés y matemáticas.</p>

<p>Esta diferencia en el promedio que hace que no sea tan alto a pesar de que haya mayores puntajes altos en inglés, es probable que se deba a motivos sociales, por lo cual, es muy importante reforzar materias como inglés en el país en general.</p>
