## Tabla de promedio entre nivel de educaci√≥n del padre con el promedio del puntaje del estudiante.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
df = pd.read_csv('https://www.datos.gov.co/resource/kgxf-xxbe.csv')

# Convertir 'puntaje_global' a num√©rico y eliminar NaN
df['punt_global'] = pd.to_numeric(df['punt_global'], errors='coerce')
df = df.dropna(subset=['fami_educacionpadre', 'punt_global'])

# Agrupar por nivel educativo y calcular el promedio del puntaje global
df_grouped = df.groupby('fami_educacionpadre')['punt_global'].mean().reset_index()

# Crear la gr√°fica de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=df_grouped['fami_educacionpadre'], y=df_grouped['punt_global'])

plt.xlabel('Nivel Educativo de los Padres')
plt.ylabel('Puntaje Global Promedio del Hijo')
plt.title('Promedio del Puntaje Global seg√∫n el Nivel Educativo de los Padres')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)  


plt.show()
```

## Gr√°fico de viol√≠n sobre la relaci√≥n entre el nivel educativo del padre y el puntaje del alumno.

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x=df['fami_educacionpadre'], y=df['punt_global'])

plt.xlabel('Nivel Educativo de los Padres')
plt.ylabel('Puntaje Global del Hijo')
plt.title('Distribuci√≥n del Puntaje Global seg√∫n Nivel Educativo de los Padres')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
sns.violinplot(x=df['fami_educacionpadre'], y=df['punt_global'], palette='deep')


plt.show()
```

Como se puede observar, el nivel educativo del padre influye en el resultado y desempe√±o del estudiante.
