```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('https://www.datos.gov.co/resource/kgxf-xxbe.csv')
df
```

```python
df["punt_matematicas"] = pd.to_numeric(df["punt_matematicas"], errors="coerce")

promedios_por_periodo = df.groupby("periodo")["punt_matematicas"].mean()

tabla_promedios = df.groupby("periodo", as_index=False)["c"].mean()
tabla_promedios.rename(columns={"punt_matematicas": "promedio_matematicas"}, inplace=True)

plt.figure(figsize=(20, 16))
plt.bar(tabla_promedios["periodo"], tabla_promedios["promedio_matematicas"], color=['orange', '#ffd000'])

plt.bar(tabla_promedios["periodo"], tabla_promedios["promedio_matematicas"], color=['orange', '#ffd000'])

plt.xlabel("Periodo")  # Etiqueta para el eje X
plt.ylabel("Promedio Matemáticas")  # Etiqueta para el eje Y
plt.title("Promedio de Matemáticas por Periodo")  # Título de la gráfica

plt.xticks(tabla_promedios["periodo"])
promedios_por_periodo

plt.grid(axis="y", linestyle="--", alpha=0.7)
```

# Los puntajes en matemáticas han mejorado con el tiempo?

***Como se puede observar el puntaje de matemáticas no ha ni mejorado ni empeorado con el tiempo, este se ha mantenido en un rango regular durante el periodo del 2013 al 2022.***

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv('https://www.datos.gov.co/resource/kgxf-xxbe.csv')

# Ver las primeras filas para identificar las columnas adecuadas
print(df.head())

# Suponiendo que las columnas son 'periodo' y 'puntaje_matematicas'
plt.figure(figsize=(10, 6))
plt.scatter(df['periodo'], df['punt_matematicas'], alpha=0.5, color='b')

# Personalizar la gráfica
plt.xlabel('Periodo')
plt.ylabel('Puntaje en Matemáticas')
plt.title('Evolución de Puntajes en Matemáticas a lo Largo del Tiempo')
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
