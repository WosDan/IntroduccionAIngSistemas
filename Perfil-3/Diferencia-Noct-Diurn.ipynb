# ¿Los estudiantes de jornada nocturna tienen puntajes más bajos que los de jornada diurna? 


### En este apartado vamos a verificar si la jornada de los estudiantes tiene relación con su desempeño.

```python
import pandas as pd

df = pd.read_csv("https://www.datos.gov.co/resource/kgxf-xxbe.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python

df = pd.read_csv("https://www.datos.gov.co/resource/kgxf-xxbe.csv")

plt.figure(figsize=(10, 5))
sns.boxplot(x=df['cole_jornada'], y=df['punt_global'], palette='Set2')

plt.xlabel('Jornada')
plt.ylabel('Puntaje Global')
plt.title('Distribución del Puntaje Global según Jornada')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

```

```python
plt.figure(figsize=(10, 5))
df_grouped = df.groupby('cole_jornada')['punt_global'].mean().reset_index()
sns.barplot(x=df_grouped['cole_jornada'], y=df_grouped['punt_global'], palette='coolwarm')
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(x=df['cole_jornada'], y=df['punt_global'], palette='pastel')

plt.xlabel('Jornada')
plt.ylabel('Puntaje Global')
plt.title('Distribución del Puntaje Global según Jornada')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# ¿Los estudiantes de jornada nocturna tienen puntajes más bajos que los de jornada diurna? 


### Como se puede observar, los estudiantes de jornada nocturna presentan un menor desempeño que los estudiantes que asisten a jornadas mañaeras o incluso completas.
