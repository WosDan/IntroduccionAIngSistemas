## Tabla de promedio entre nivel de educación del padre con el promedio del puntaje del estudiante.

### Solo 1000 primeras filas (El API)

### Base de datos completa

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://www.datos.gov.co/resource/kgxf-xxbe.csv')

df['punt_global'] = pd.to_numeric(df['punt_global'], errors='coerce')
df = df.dropna(subset=['fami_educacionpadre', 'punt_global'])

df_grouped = df.groupby('fami_educacionpadre')['punt_global'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x=df_grouped['fami_educacionpadre'], y=df_grouped['punt_global'])

plt.xlabel('Nivel Educativo de los Padres')
plt.ylabel('Puntaje Global Promedio del Hijo')
plt.title('Promedio del Puntaje Global según el Nivel Educativo de los Padres')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)  


plt.show()
```

## Gráfico de violín sobre la relación entre el nivel educativo del padre y el puntaje del alumno.

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x=df['fami_educacionpadre'], y=df['punt_global'])

plt.xlabel('Nivel Educativo de los Padres')
plt.ylabel('Puntaje Global del Hijo')
plt.title('Distribución del Puntaje Global según Nivel Educativo de los Padres')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
sns.violinplot(x=df['fami_educacionpadre'], y=df['punt_global'], palette='pastel')


plt.show()
```

## Tabla de promedio entre nivel de educación de la madre con el promedio del puntaje del estudiante.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# df = pd.read_csv('https://www.datos.gov.co/resource/kgxf-xxbe.csv')

df['punt_global'] = pd.to_numeric(df['punt_global'], errors='coerce')
df = df.dropna(subset=['fami_educacionmadre', 'punt_global'])

df_grouped = df.groupby('fami_educacionmadre')['punt_global'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x=df_grouped['fami_educacionmadre'], y=df_grouped['punt_global'])

plt.xlabel('Nivel Educativo de las Madres')
plt.ylabel('Puntaje Global Promedio del Hijo')
plt.title('Promedio del Puntaje Global según el Nivel Educativo de las Madres')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)  


plt.show()
```

## Gráfico de violín sobre la relación entre el nivel educativo de la Madre y el puntaje del Alumno.

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x=df['fami_educacionmadre'], y=df['punt_global'])

plt.xlabel('Nivel Educativo de las Madres')
plt.ylabel('Puntaje Global del Hijo')
plt.title('Distribución del Puntaje Global según Nivel Educativo de las Madres')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
sns.violinplot(x=df['fami_educacionmadre'], y=df['punt_global'], palette='deep')


plt.show()
```

<p>De acuerdo a las gráficas podemos observar que a mayor es el nivel educativo mayor promedio global obtienen los estudiantes, esto significa que el conocimiento en ciertos aspectos de los padres puede afectar el desempeño general del niño, resultando en mayores oportunidades para este si sus padres llegaron a niveles superiores de educación y si la completaron.</p>
